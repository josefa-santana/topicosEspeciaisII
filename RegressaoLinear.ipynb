```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mlxtend.plotting import scatterplotmatrix
```

```python
mpg=sns.load_dataset('mpg')
sns.set(style='whitegrid', context='notebook')
mpg.head()
```

```python
cols = ['mpg', 'displacement', 'horsepower', 'acceleration']
scatterplotmatrix(mpg[cols].values, figsize=(10, 8),
names=cols, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set(style='whitegrid', context='notebook')
cols = ['mpg', 'displacement', 'horsepower', 'acceleration']
sns.pairplot(mpg[cols], height=2.5)
plt.savefig('scatter.png', dpi=300)
plt.show()
```

```python
corr_cols = ['mpg', 'displacement', 'horsepower', 'acceleration']
```

```python
cm = np.corrcoef(mpg[corr_cols].values.T)
sns.set(font_scale=1.5)
fig, ax = plt.subplots(figsize=(14, 14))
hm = sns.heatmap(cm,
ax=ax,
cbar=False,
annot=True,
square=True,
fmt='.2f',
annot_kws={'size': 15},
yticklabels=corr_cols,
xticklabels=corr_cols)
plt.show()
```
