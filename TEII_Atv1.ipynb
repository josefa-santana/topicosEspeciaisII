```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.exponential(scale=1.5, size=1000)
```

```python
plt.rcParams.update({'font.size': 12})

plt.figure()
plt.title("Distribuição de probabilidade exponencial")
plt.plot(x,'.-')
plt.xlabel('qual dado')
plt.ylabel('valores')
# faz um histograma dos valores

plt.figure()
plt.hist(x)
plt.xlabel('valores')
plt.ylabel('frequência')
```
