# Exercício 1

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.exponential(scale=1.5, size=1000)
```

```python
# gera um gráfico dos valores
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.title("Distribuição de probabilidade exponencial")
plt.plot(x,'.-')
plt.xlabel('qual dado')
plt.ylabel('valores')
```

```python
# faz um histograma dos valores
plt.figure()
plt.title("Distribuição de probabilidade exponencial")
plt.hist(x)
plt.xlabel('valores')
plt.ylabel('frequência')
```

# Exercício 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
A=0;
B=1;
P_AB = expon.cdf(B)-expon.cdf(A)
print(f'prob entre {A:.1f} e {B:.1f} = {P_AB:.3f}')
```

```python
bins=np.linspace(-5,5,1000)
gauss = expon.pdf(bins, 0, 1)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(bins, gauss, 'b--', linewidth=1)
plt.xlabel('valores')
plt.ylabel('distribuição de probabilidade exponencial')
px=bins[(bins>A)&(bins<B)]
plt.fill_between(px,gauss[(bins>A)&(bins<B)],color='k',alpha=0.2)
```
